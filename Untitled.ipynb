```r
df <- data.frame(x=c(2,8,9,10),y=c(4,15,8,13))
```

```r
library(ggplot2)
library(dplyr)
library(caret)
```

```r
model_ls <- function(x) 1.032*x + 2.516
model_xls <- function(x) 1.023*x + 2.663
model_xls2 <- function(x) 1.017*x + 1.846
```

```r
df %>% 
ggplot(aes(x,y))+
geom_point()+
stat_function(fun = model_ls,color='blue')+
stat_function(fun = model_xls,color='red')+
stat_function(fun = model_xls2,color='orange')
```

```r
df %>% 
mutate(y_hat_ls = model_ls(x),y_hat_xls = model_xls(x),
       y_hat_xls2 = model_xls2(x)) -> df_2


R2(df_2$y_hat_ls,df_2$y) - R2(df_2$y_hat_xls2,df_2$y)
```
